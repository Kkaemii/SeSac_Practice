```python
city = ['서울', '인천', '대전', '대구', '울산', '부산', '광주'] 
# 위도(Latitude)와 경도(Longitude) 
lat = [37.56, 37.45, 36.35, 35.87, 35.53, 35.18, 35.16]
lon = [126.97, 126.70, 127.38, 128.60, 129.31, 129.07, 126.85] 
# 인구 밀도(명/km^2): 2017년 통계청 자료 
pop_den = [16154, 2751, 2839, 2790, 1099, 4454, 2995] 


size= np.array(pop_den)  * 0.2 # array로 각 배열에 크기를 다르게 설정하기 위해서 작업
colors = ['r', 'g', 'b', 'c', 'm', 'k', 'y'] #마커의 색상 지정

plt.scatter (lon, lat, s=size, c=colors, alpha=0.5) 
plt.xlabel('경도 (longitude)') 
plt.ylabel('위도 (latitude)') 
plt.title('지역별 인구 밀도(2017)') 

for x, y, name in zip(lon, lat, city): 
	plt.text(x, y, name) # 위도와 경도에 맞게 도시 이름 출력

plt.show()
```

```python
member_IDs = ['m_01','m_02','m_03','m_04']
before_ex =[27,35,40,33]
after_ex =[30,38,42,37]

n_data=len(member_IDs) #데이터의 개수
index = np.arange(1,n_data+1) #데이터의 개수만큼 인덱스 생성
plt.bar(index, before_ex)
plt.show()
```

```python
grade_num=[5,14,12,3]
students=['A','B','C','D']

dict_data = {'학생': students, '점수':grade_num}
df_grade = pd.DataFrame(dict_data, columns=['학생','점수'])
df_grade.plot.bar(x='학생', y='점수', grid=True, title='학점별 학생 수 막대 그래프', rot=0)
plt.show()
```

```python
grade_num=[5,14,12,3]
students=['A','B','C','D']

dict_data = {'학생': students, '점수':grade_num}
df_grade = pd.DataFrame(dict_data, columns=['학생','점수'])
df_grade.plot.bar(x='학생', y='점수', grid=True, title='학점별 학생 수 막대 그래프', rot=0)
```

```python
grade_num=[5,14,12,3]
students=['A','B','C','D']
df_grade = pd.DataFrame(grade_num, index=students, columns=['Student'])
df_grade.plot.bar(rot=0,grid=True).set_xlabel('학점')
df_grade.plot.bar(rot=0,grid=True).set_ylabel('학점')
df_grade.plot.bar(rot=0,grid=True).set_title('학점별 학생 수 막대 그래프')
plt.show()
```
