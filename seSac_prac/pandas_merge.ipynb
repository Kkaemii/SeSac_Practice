```python
import seaborn as sns

tips = sns.load_dataset('tips')
tips
```

```python
tips.info()
tips.day.value_counts()
```

```python
tips.value_counts('sex') #성별 인원수
```

```python
#흡연 유무에 따른 성별 인원
smoke_=tips.groupby(['sex','smoker'], observed=False).size().reset_index(name='count')
smoke_[smoke_['sex']=='Male']
smoke_[smoke_['sex']=='Female']
```

```python
t_df = tips.groupby(['sex', 'smoker'],observed=False).size().reset_index(name='counter')

newDf = pd.pivot_table(tips, index='smoker',columns='sex',values='tip',aggfunc='count',observed=False)
newDf
```

```python
#성별에 따른 팁의 평균
re_tip1 = tips.groupby('sex', observed=False)
re_tip1
re_tip2 = tips.groupby('sex',observed=False)['tip'].mean()
re_tip2
```

```python
#흡연 유무에 따른 팁 평균
tips.groupby('smoker', observed=False)['tip'].mean()
```

```python
tips.groupby('day',observed=False).size().reset_index(name='people_count')
```

```python
tips.groupby(['day','sex'],observed=False).size().reset_index(name='count')
```

```python
tips.groupby('day',observed=False)['size'].mean().reset_index(name='평균 인원수')
```

```python
tips.groupby('day', observed=False)['tip'].mean().reset_index(name='평균 팁')
```
