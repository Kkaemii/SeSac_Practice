```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#벅스 뮤직
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Trident/7.0; rv:11.0) like Gecko'}
req = requests.get('https://music.bugs.co.kr/chart', headers = header)
html = req.text
bugs = BeautifulSoup(html, 'lxml')
```

```python
b_rank = bugs.find_all(class_='ranking')
b_title = bugs.find_all('p', {'class':'title'})
b_artist = bugs.find_all('p', {'class':'artist'})


b_rank_result = []
b_title_result = []
b_artist_result = []

try:
	#순위 리스트 만들기
	print("랭킹 리스트  만드는 중")
	for rank in range(1, len(b_rank)):
		b_rank_result.append(b_rank[rank].find('strong').get_text())
		
	#노래 제목 리스트 만들기
	print("제목 리스트 만드는 중")
	for title in range(len(b_title)):
		b_title_result.append(b_title[title].get_text().replace('\n',''))

	#가수 리스트 만들기
	print("가수 리스트 만드는 중")
	for singer in range(len(b_artist)):
		b_artist_result.append(b_artist[singer].get_text().replace('\n',''))
except Exception as e:
	print(f"에러 발생: {e.__traceback__.tb_lineno}")
	print(f"{e} 에러가 발생했습니다.")

# #rank 인덱스 1번부터 시작해야함. 0번은 순위나옴
# print(b_rank[2].find('strong').get_text())
# print(b_title[1].get_text().replace('\n',''))
# print(b_artist[1].get_text().replace('\n',''))
```

```python
#벅스_데이터 프레임으로 만들기

b_data = {'순위':b_rank_result, '가수명' : b_artist_result, '노래제목': b_title_result}
bugs_df = pd.DataFrame(b_data)
#requires openpyxl package
#df is your dataframe
bugs_df.to_excel("./벅스뮤직 Top100.xlsx", sheet_name='SheetName', index=False)
```

```python
#멜론
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Trident/7.0; rv:11.0) like Gecko'}
req = requests.get('https://www.melon.com/chart/index.htm', headers = header)
html = req.text
melon = BeautifulSoup(html, 'lxml')
```

```python
m_rank = melon.find_all('span',{'class':'rank'})
m_title = melon.find_all('div',{'class':'ellipsis rank01'})
m_artist = melon.find_all('span',{'class' : 'checkEllipsis'},'a')

#인덱스 넘버 바꿔서 진행하면 됨
#0번 인덱스가 순위라서 1번인덱스 부터해야 1위로 출력됨

m_rank_result=[]
m_title_result=[]
m_artist_result=[]

# print(m_rank[1].get_text()) 
# print(m_title[0].get_text().replace('\n',''))
# print(m_artist[0].get_text())

try:
	#순위 리스트 만들기
	print("랭킹 리스트  만드는 중")
	for rank in range(1, len(m_rank)):
		m_rank_result.append(m_rank[rank].get_text())
		
	#노래 제목 리스트 만들기
	print("제목 리스트 만드는 중")
	for title in range(len(m_title)):
		m_title_result.append(m_title[title].get_text().replace('\n',''))

	#가수 리스트 만들기
	print("가수 리스트 만드는 중")
	for singer in range(len(m_artist)):
		m_artist_result.append(m_artist[singer].get_text())
except Exception as e:
	print(f"에러 발생: {e.__traceback__.tb_lineno}")
	print(f"{e} 에러가 발생했습니다.")
```

```python
#멜론_데이터 프레임으로 만들기

m_data = {'순위':m_rank_result, '가수명' : m_artist_result, '노래제목': m_title_result}
melon_df = pd.DataFrame(m_data)
#requires openpyxl package
#df is your dataframe
melon_df.to_excel("./멜론 Top100.xlsx", sheet_name='SheetName', index=False)
```

```python
import time

#지니뮤직
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Trident/7.0; rv:11.0) like Gecko'}

#1위부터 100위 까지
g_url = ['https://www.genie.co.kr/chart/top200?&pg=1','https://www.genie.co.kr/chart/top200?&pg=2' ]

genie_pages=[]
for url in g_url:
	req = requests.get(url, headers = header)
	time.sleep(3)
	genie = BeautifulSoup(req.text, 'lxml')
	genie_pages.append(genie)

#지니뷰직은 페이지 2개로 해야 100개 받을 수있음
# 두번째 페이지 url https://www.genie.co.kr/chart/top200&pg=2
```

```python


#2페이지에 나뉜 데이터 하나로 통합하기
#extend를 쓰면 리스트 안에 요소를 넣는것으로 2중리스트 안됨

#지나 차트가 상위에 1~5위 항목을 보여주는 섹션이 있어, 우리가 원하는 리스트가 담겨있는 섹션을 따로 선택하여 진행한다.


g_rank_all=[]
g_title_all=[]
g_artist_all=[]




for i in range(2):
	genie_body = genie_pages[i].select_one('.list-wrap > tbody') #서치하는 범위를 지정
	g_rank_all.extend(genie_body.find_all('td',{'class':'number'}))
	g_title_all.extend(genie_body.find_all('a',{'class':'title ellipsis'}))
	g_artist_all.extend(genie_body.find_all('a',{'class':'artist ellipsis'}))



#최종 결과를 담을 리스트 생성
g_rank_result=[]
g_title_result=[]
g_artist_result=[]


try:
	#순위 리스트 만들기
	print("랭킹 리스트  만드는 중")
	for rank in range(len(g_rank_all)): #get_text()의 결과는 문자열로, 리스트에 담아도 2중리스트 x
		g_rank_result.append(g_rank_all[rank].get_text().replace('\n','')[:3].strip())
		
	#노래 제목 리스트 만들기
	print("제목 리스트 만드는 중")
	for title in range(len(g_title_all)):
		g_title_result.append(g_title_all[title].get_text().replace('\n','').strip())

	#가수 리스트 만들기
	print("가수 리스트 만드는 중")
	for singer in range(len(g_artist_all)):
		g_artist_result.append(g_artist_all[singer].get_text())
except Exception as e:
	print(f"에러 발생: {e.__traceback__.tb_lineno}")
	print(f"{e} 에러가 발생했습니다.")
```

```python
#지니_데이터 프레임으로 만들기

g_data = {'순위':g_rank_result, '가수명' : g_artist_result, '노래제목': g_title_result}
genie_df = pd.DataFrame(g_data)
genie_df.to_excel("./지니뮤직 Top100.xlsx", sheet_name='SheetName', index=False)
```

```python
result = pd.concat([bugs_df, melon_df, genie_df] ,axis=1)
```

```python
result.columns = pd.MultiIndex.from_tuples([
		('순위', ''),	('벅스 뮤직', '가수명'), ('벅스 뮤직', '노래제목'),
		('순위', 'a'),('멜론 뮤직', '가수명'), ('멜론 뮤직', '노래제목'),#추후에 중복 열 제거를 위해 a,b 대입
		('순위', 'b'),('지니 뮤직', '가수명'), ('지니 뮤직', '노래제목')
    ])
```
