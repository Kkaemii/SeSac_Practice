## read_sql( )

SQL 연결 객체를 활용하여 쿼리 구문으로 반환된 결과를 데이터프레임으로

```python
import pymysql
#db 연결하기
db = pymysql.connect(host='localhost', port=3306,user='root',passwd='1234', charset='utf8')
#커서 가지고오기
cursor = db.cursor()
```

```python
#student_mgmt 데이터 베이스 만들기

sql='CREATE DATABASE IF NOT EXISTS student_mgmt;'
cursor.execute(sql)

#데이터 베이스 선택
sql = 'USE student_mgmt;'
cursor.execute(sql)
```

하나의 문자열에 여러 sql문을 포함하고있다면 실행이 되지 않는다.
각 sql문을 단독으로 실행시키거나 client_flag 매개변수를 사용할 수 있다.

import pymysql
from pymysql.constants import CLIENT

connection = pymysql.connect(host='localhost', user='username', password='password', db='database_name',
                             client_flag=CLIENT.MULTI_STATEMENTS)

```python

#테이블 검증 (만약 동일한 테이블 있다면 drop하고 다시 만듦)
cursor.execute('DROP TABLE IF EXISTS students;')

#테이블 만들기
sql ="""
	CREATE TABLE students (
	id TINYINT NOT NULL AUTO_INCREMENT,
	name VARCHAR(10) NOT NULL,
	gender ENUM('man','woman') NOT NULL,
	birth DATE NOT NULL,
	english TINYINT NOT NULL,
	math TINYINT NOT NULL,
	korean TINYINT NOT NULL,
	PRIMARY KEY (id)
	) ENGINE=InnoDB DEFAULT CHARSET=utf8;
"""
cursor.execute(sql)
```

```python
#데이터 입력 쿼리
#pymysql 동작특성상 다중 쿼리가 실행되지 않아서 각자 개별 작동 코드 작성

"""
INSERT INTO students (name, gender, birth, english, math, korean) VALUES ('dave', 'man', '1983-07-16', 90, 80, 71);
INSERT INTO students (name, gender, birth, english, math, korean) VALUES ('minsun', 'woman', '1982-10-16', 30, 88, 60);
INSERT INTO students (name, gender, birth, english, math, korean) VALUES ('david', 'man', '1982-12-10', 78, 77, 30);
INSERT INTO students (name, gender, birth, english, math, korean) VALUES ('jade', 'man', '1979-11-1', 45, 66, 20);
INSERT INTO students (name, gender, birth, english, math, korean) VALUES ('jane', 'man', '1990-11-12', 65, 32, 90);
INSERT INTO students (name, gender, birth, english, math, korean) VALUES ('wage', 'woman', '1982-1-13', 76, 30, 80);
INSERT INTO students (name, gender, birth, english, math, korean) VALUES ('tina', 'woman', '1982-12-3', 87, 62, 71);
"""
cursor.execute("INSERT INTO students (name, gender, birth, english, math, korean) VALUES ('dave', 'man', '1983-07-16', 90, 80, 71);")
cursor.execute("INSERT INTO students (name, gender, birth, english, math, korean) VALUES ('minsun', 'woman', '1982-10-16', 30, 88, 60);")
cursor.execute("INSERT INTO students (name, gender, birth, english, math, korean) VALUES ('david', 'man', '1982-12-10', 78, 77, 30);")
cursor.execute("INSERT INTO students (name, gender, birth, english, math, korean) VALUES ('jade', 'man', '1979-11-1', 45, 66, 20);")
cursor.execute("INSERT INTO students (name, gender, birth, english, math, korean) VALUES ('jane', 'man', '1990-11-12', 65, 32, 90);")
cursor.execute("INSERT INTO students (name, gender, birth, english, math, korean) VALUES ('wage', 'woman', '1982-1-13', 76, 30, 80);")
cursor.execute("INSERT INTO students (name, gender, birth, english, math, korean) VALUES ('tina', 'woman', '1982-12-3', 87, 62, 71);")
```

```python
import pandas as pd
import pymysql

def conn(d_name):
  import pymysql
  db = pymysql.connect(host='localhost', port=3306, user='root', passwd='1234', charset='utf8', db =d_name )
  return db

#데이터베이스 이름이 로 데이터베이스에 연결
db = conn('sqldb')
```

```python
cursor = db.cursor()
sql = "INSERT INTO userTbl VALUES('STJ', '서태지', 1975, '경기', '011', 'toortoor', 171, '2014-4-4');"
cursor.execute(sql)
db.commit()
```

```python
sql = "INSERT INTO buyTbl (userID, prodName, groupName, price, amount) VALUES('STJ', '운동화', '의류', 30, 2);"
cursor.execute(sql)
db.commit()
```

```python
#외래키로 참조하고있기때문에 
sql_query = "delete from userTbl where userID='STJ'"
cursor.execute(sql_query)
db.commit()
```
